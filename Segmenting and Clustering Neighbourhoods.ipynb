Next I read the Wikipedia table into a dataframe.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df = pd.read_html(url)[0]

df.head()
```

Then I removed all rows from the 'Borough' column which had the value 'Not assigned'.

```python
df_seg = df[df.Borough != 'Not assigned']
df_seg.head(10)
```

The next step was to replace all 'Not assigned' values of the 'Neighbourhood' column with their respective 'Borough' values.

```python
df_seg['Neighbourhood'].replace('Not assigned',df_seg['Borough'],inplace=True)
df_seg
```

After that, I grouped the 'Neighbourhood' column by 'Postcode' and then 'Borough'.  This put all values from 'Neighbourhood' into the same row if they shared the same post code.

```python
df_seg['Neighbourhood'].groupby([df_seg.Postcode,df_seg.Borough]).apply(list).reset_index()
```
